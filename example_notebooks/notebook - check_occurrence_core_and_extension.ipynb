### Notebook to test Pyobistools' function 'check_occurrence_core_and_extension'

##### Function 'check_occurrence_core_and_extension' description
The 'check_occurrence_core_and_extension' function is used with one file of either Darwin Core file types 'occurrence_core' or 'occurrence_extension', to report:
+ duplicates values in 'occurrenceID'
+ check if values of field 'occurrenceStatus' are either 'present' or 'absent'
+ check if values of field 'basisOfRecord' belongs (case-sensitive) to the following list ['PreservedSpecimen', 'FossilSpecimen', 'LivingSpecimen', 'HumanObservation', 'MachineObservation', 'MaterialSample', 'MaterialCitation', 'MaterialEntity', 'Occurrence', 'Taxon', 'Event']

##### Function 'check_occurrence_core_and_extension' arguments
+ data: Dataframe of the data to evaluate
